```python
import json

points_data_json = "./raw_data.json"
points_output_json = "./points_output.json"

with open(points_data_json, "r") as f:
    batch_data = json.load(f)
    
with open(points_output_json, "r") as f:
    prefix_data = json.load(f)
```

```python
print(batch_data)
print(prefix_data)
```

```python
#  prefix 
prefix_points_list = []

for node in prefix_data:
    prefix = node["prefix"]
    indices = node["points"]  # 
    matched_pts = []
    matched_norm_pts = []

    for idx in indices:
        # idx  batch_data 
        matched_pts.append(batch_data[0][idx]["point"])        #  pts
        matched_norm_pts.append(batch_data[0][idx]["norm_point"])  #  pts

    prefix_points_list.append({
        "prefix": prefix,
        "pts": matched_pts,
        "norm_pts": matched_norm_pts
    })
```

```python
#  prefix
target_prefix = 3

#  prefix 
node = next((n for n in prefix_data if n["prefix"] == target_prefix), None)

if node is not None:
    indices = node["points"]  #  batch 
    matched_pts = [batch_data[0][idx]["point"] for idx in indices]
    matched_morton_codes = [batch_data[0][idx]["morton_code"] for idx in indices]

    print(f"Prefix {target_prefix} points:", matched_pts)
    print(f"Prefix {target_prefix} morton codes:", matched_morton_codes)
else:
    print(f"Prefix {target_prefix} not found!")
```
